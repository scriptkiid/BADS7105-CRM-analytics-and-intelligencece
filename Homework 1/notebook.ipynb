```python
import pandas as pd
```

# Read Input

```python
input = pd.read_csv("./1 Raw Data/Customer Behaviors (Responses).csv")
input.head(3)
```

# Cleansing Data

```python
cleansing_data = input.copy()

for key in input.keys():
    if "คุณมีความสนใจในสิ่งเหล่านี้มากน้อยเพียงใด" in key:
        cleansing_data[key] = input[key].replace({
            "ไม่สนใจอย่างมากที่สุด": 0,
            "ไม่สนใจอย่างมาก": 1,
            "ไม่สนใจ": 2,
            "เฉยๆ": 3,
            "สนใจ": 4,
            "สนใจอย่างมาก": 5,
            "สนใจอย่างมากที่สุด": 6
        })
    elif "คุณบริโภคสิ่งเหล่านี้บ่อยขนาดไหน" in key:
        cleansing_data[key] = input[key].replace({
            "แทบไม่ได้บริโภคเลย": 0,
            "หลายเดือนครั้ง": 1,
            "เดือนละครั้ง": 2,
            "เดือนละ 2-3 ครั้ง": 3,
            "อาทิตย์ละครั้ง": 4,
            "แทบทุกวัน": 5
        })

cleansing_data.head(5)
```

# Normalize Data

```python
from sklearn.preprocessing import MinMaxScaler

normalize_data = cleansing_data.iloc[:, 1:61]
scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(normalize_data) 
normalize_data.loc[:,:] = scaled_values
normalize_data["เพศ"] = cleansing_data["เพศ"]
normalize_data.head(5)
```

```python
data = []

for key in normalize_data.iloc[:, :60].keys():
    data += [[
        category[key[key.index("["): len(key)]],
        key,
        normalize_data[key].mean(),
        normalize_data.loc[normalize_data["เพศ"] == "ชาย"][key].mean(),
        normalize_data.loc[normalize_data["เพศ"] == "หญิง"][key].mean(),
        normalize_data.loc[normalize_data["เพศ"] == "ไม่ต้องการระบุ"][key].mean()
    ]]

mean_data = pd.DataFrame(data, columns = ["Category", "Variable", "Mean Total", "Mean Male", "Mean Female", "Mean Other"])
mean_data[0: 5]
```

# Correlation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = normalize_data.iloc[:,1:61].corr()
correlation_matrix.index = range(59)
correlation_matrix.columns = range(59)
sns.heatmap(correlation_matrix)

fig = plt.gcf()
fig.set_size_inches(20,16)
plt.show()
```
