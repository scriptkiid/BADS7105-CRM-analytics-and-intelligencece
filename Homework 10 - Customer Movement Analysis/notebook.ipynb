```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result_df = pd.read_csv('results.csv')
```

```python
fig, ax = plt.subplots(1, figsize=(16, 8))
plt.bar(result_df.index, result_df['NEW_CUSTOMER'], bottom=[i+j for i,j in zip(result_df['REACTIVATE_CUSTOMER'], result_df['REPEAT_CUSTOMER'])], color = '#0a2549', width =0.5)
plt.bar(result_df.index, result_df['REACTIVATE_CUSTOMER'], bottom=result_df['REPEAT_CUSTOMER'], color = '#104681', width =0.5)
plt.bar(result_df.index, result_df['REPEAT_CUSTOMER'], color = '#3d9be9', width =0.5)
plt.bar(result_df.index, result_df['CHURN_CUSTOMER'], color = '#f24256', width =0.5)

# x and y limits
plt.ylim(-1300, 1800)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

#grid
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed', alpha=0.7)

# x ticks
xticks_labels = result_df['CURRENT_YEAR_MONTH'].tolist()
plt.xticks(result_df.index, labels = xticks_labels, rotation=75)

# title and legend
legend_label = ['NEW_CUSTOMER', 'REACTIVATE_CUSTOMER', 'REPEAT_CUSTOMER', 'CHURN_CUSTOMER']
plt.legend(legend_label, bbox_to_anchor=([1, 1.05, 0, 0]), frameon = False)
plt.show()
```
